You will learn here how to generate a 1D Gaussian random field and you will look at the effect of a Fourier mode

```python
# number of cells (power of two is usually good)
N=2**6
# number of Fourier modes
NF=int(N/2+1)
# length of array, and cell width
dx = 1
L = dx * N
# random seed
seed=1234
# setup random number generator using seed
gen = np.random.RandomState(seed)
```

### Input power spectrum

The power spectrum will deteremine how much structure is there as a funciton of scale

```python
def power(k,k0=0.05):
    """Define here your favorite power spectrum. Mine is a Gaussian around k0."""
    P = np.zeros_like(k)
    # avoid log(k=0)
    nonzero = (k>0)
    P[nonzero] = np.exp(-np.log(k[nonzero]/k0)**2)
    return P
```

```python
if show_plot:
    kmin=2*np.pi/L
    kmax=np.pi/dx
    k=np.logspace(np.log10(kmin),np.log10(kmax),1000)
    P_01=power(k,0.1)
    P_02=power(k,0.2)
    P_05=power(k,0.5)
    plt.xlim(kmin,kmax)
    plt.xlabel('k',fontsize=15)
    plt.ylabel('P(k)',fontsize=15)
    plt.semilogx(k,P_01,label='k0=0.1')
    plt.semilogx(k,P_02,label='k0=0.2')
    plt.semilogx(k,P_05,label='k0=0.5')
    plt.legend()
```

### Generate random Fourier modes (complex numbers)

```python
def get_gaussian_modes(k0=1.0,new_seed=None):
    """Generate Gaussian random numbers for the Fourier modes of the grid.
    
    k0 is used by power spectrum, new_seed is useful to ensure same seed in fields"""
    if new_seed:
        gen = np.random.RandomState(new_seed)
        
    # get frequencies (wavenumbers)
    k = np.fft.rfftfreq(N)*2*np.pi/dx
    # get power evaluated at each k
    P = power(k,k0)

    # generate random Fourier modes (1 field only)
    Ns=1
    modes = np.empty([Ns,NF], dtype=complex)
    modes[:].real = np.reshape(gen.normal(size=Ns*NF),[Ns,NF])
    modes[:].imag = np.reshape(gen.normal(size=Ns*NF),[Ns,NF])
    # normalize to desired power (and enforce real for i=0, i=NF-1)
    modes[:,0] = modes[:,0].real * np.sqrt(P[0])
    modes[:,-1] = modes[:,-1].real * np.sqrt(P[-1])
    modes[:,1:-1] *= np.sqrt(0.5*P[1:-1])    
    return modes[0]
    
def get_field(modes):
    """Given a set of Fourier modes, compute inverse Fourier transform to get a field"""
    if NF != len(modes):
        print('size mismatch',NF,len(modes))
        raise SystemExit
    # inverse FFT to get (normalized) delta field
    delta = np.fft.irfft(modes) * np.sqrt(N/dx)
    return delta

def get_modes(delta):
    """Given a delta field, return set of Fourier modes using FFT"""
    if N != len(delta):
        print('size mismatch',N,len(delta))
        raise SystemExit
    # inverse FFT to get (normalized) delta field
    modes = np.fft.rfft(delta) / np.sqrt(N/dx)
    new_delta = np.fft.irfft(modes) * np.sqrt(N/dx)
    return modes

def get_gaussian_field(k0=0.01,new_seed=None):
    """Generate a set of random Fourier modes, and use them to get a field"""
    modes = get_gaussian_modes(k0,new_seed)
    return get_field(modes)
```

### Study the effect of input power

```python
# get Gaussian fields for a couple of different power spectrum (different k0), but same seed!
modes_01 = get_gaussian_modes(0.1,seed)
delta_01 = get_field(modes_01)
modes_05 = get_gaussian_modes(0.5,seed)
delta_05 = get_field(modes_05)
modes_02 = get_gaussian_modes(0.2,seed)
delta_02 = get_field(modes_02)
```

```python
x = dx*np.arange(N)
if show_plot:
    plt.plot(x,delta_05,label='k0=0.5')
    plt.plot(x,delta_02,label='k0=2.0')
    plt.plot(x,delta_01,label='k0=1.0')
    plt.xlim(0.0,100.0)
    plt.xlabel('x')
    plt.ylabel('Gaussian field')
    plt.legend()
```

### Check agreement with input power

The squared amplitude of the Fourier modes should (on average!) agree with the input power spectrum

```python
k = np.fft.rfftfreq(N)*2*np.pi/dx
if show_plot:
    plt.plot(k,abs(modes_05)**2,label='k0=0.5',color='red')
    plt.plot(k,abs(modes_02)**2,label='k0=0.2',color='blue')
    plt.plot(k,abs(modes_01)**2,label='k0=0.1',color='black')
    plt.plot(k,power(k,0.5),color='red',ls=':')
    plt.plot(k,power(k,0.2),color='blue',ls=':')    
    plt.plot(k,power(k,0.1),color='black',ls=':')
    plt.xlabel('k')
    plt.ylabel('P(k)')
    plt.legend()
```

```python
def binned_power(k,Pk,Nrebin):
    if NF != len(Pk):
        print('size mismatch',NF,len(Pk))
        raise SystemExit
    Nk=int(NF/Nrebin)
    #print('NF',NF)
    #print('Nk',Nk)
    binned_k = numpy.zeros(Nk)
    binned_Pk = numpy.zeros(Nk)
    binned_var_Pk = numpy.zeros(Nk)
    for i in range(0,NF):
        ibin=int(i/Nrebin)
        if ibin==Nk:
            continue
        binned_k[ibin] += k[i]
        binned_Pk[ibin] += Pk[i]
        binned_var_Pk[ibin] += Pk[i]*Pk[i] 
    for ibin in range(0,Nk):
        binned_k[ibin] /= Nrebin
        binned_Pk[ibin] /= Nrebin
        # variance in binned power
        binned_var_Pk[ibin] /= (Nrebin-1)
        binned_var_Pk[ibin] -= binned_Pk[ibin]*binned_Pk[ibin]
        # error in mean of binned power
        binned_var_Pk[ibin] /= Nrebin
    return binned_k, binned_Pk, binned_var_Pk
```

```python
Nrebin=2
bin_k,bin_Pk_05,bin_var_Pk_05 = binned_power(k,abs(modes_05)**2,Nrebin)
bin_k,bin_Pk_02,bin_var_Pk_02 = binned_power(k,abs(modes_02)**2,Nrebin)
bin_k,bin_Pk_01,bin_var_Pk_01 = binned_power(k,abs(modes_01)**2,Nrebin)

if show_plot:   
    plt.errorbar(bin_k,bin_Pk_05,yerr=np.sqrt(bin_var_Pk_05),fmt='.',label='k0=0.5',color='red')
    plt.errorbar(bin_k,bin_Pk_02,yerr=np.sqrt(bin_var_Pk_02),fmt='.',label='k0=0.2',color='blue')
    plt.errorbar(bin_k,bin_Pk_01,yerr=np.sqrt(bin_var_Pk_01),fmt='.',label='k0=0.1',color='black')
    plt.plot(k,power(k,0.5),color='red',ls=':')
    plt.plot(k,power(k,0.2),color='blue',ls=':')    
    plt.plot(k,power(k,0.1),color='black',ls=':')
    plt.xlabel('k')
    plt.ylabel('P(k)')
    plt.legend()
```

### Mess around with a single Fourier mode
In order to show the role of a given Fourier mode, we will mess around with the sixth lower-k mode.
We will start by boosting the amplitude of the mode.

```python
modes=np.copy(modes_01)
delta=get_field(modes)
modes_boost=np.copy(modes)
boost=5
modes_boost[5] = boost*modes_boost[5]
if show_plot:
    plt.plot(k,abs(modes),label='original')
    plt.plot(k,abs(modes_boost),label='boost')
    plt.xlabel('k')
    plt.ylabel('Gaussian modes')
    plt.legend()
```

```python
delta_boost=get_field(modes_boost)
if show_plot:
    plt.plot(x,delta,label='original')
    plt.plot(x,delta_boost,label='boost')
    plt.xlabel('x')
    plt.ylabel('Gaussian field')
    plt.legend()
```

We can also vary the phase of the Fourier mode, without changing the overall amplitude (it is a complex number!)

```python
abs_5=boost*abs(modes[5])
angle_5=angle(modes[5])

modes_0=np.copy(modes)
modes_90=np.copy(modes)
modes_180=np.copy(modes)
modes_270=np.copy(modes)

modes_0[5] = abs_5 * exp(1j*angle_5)
modes_90[5] = abs_5 * exp(1j*(angle_5+0.5*np.pi))
modes_180[5] = abs_5 * exp(1j*(angle_5+np.pi))
modes_270[5] = abs_5 * exp(1j*(angle_5+1.5*np.pi))

delta_0=get_field(modes_0)
delta_90=get_field(modes_90)
delta_180=get_field(modes_180)
delta_270=get_field(modes_270)
```
